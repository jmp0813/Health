```python
class DecimalScaler :
    def __init__(self) :
        self.dec = 0
        return
    
    def calculate(self, data) :
        return data / self.dec
    
    def fit(self, data) :
        self.dec = 10 ** np.array([len(i.split('.')[0]) for i in np.floor(data.max()).astype(str)])
        return
        
    def fit_transform(self, data) :
        self.fit(data)
        return self.calculate(data)
    
    def transform(self, data) :
        return self.calculate(data)
    
    def inverse_transform(self, data) :
        return data * self.dec
```

```python
time_length = 600
n_feature = 6

def makeTimeData(augmentData, time_length, col_names) :
    originalFeatureData = np.empty(shape = (augmentData["id"].unique().shape[0], time_length, len(col_names)))

    i = 0
    for idx in tqdm.tqdm(augmentData.index.unique()) :
        originalFeatureData[i] = augmentData.loc[idx, col_names]
        i += 1
    return originalFeatureData
```

```python
# Under class data Oversampling

rawFeature = pd.read_csv("./RawData/train_features.csv")
rawTarget = pd.read_csv("./RawData/train_labels.csv")
rawFeature.index = rawFeature["id"]

axisComb = [
    ["id", "time", "acc_x", "acc_z", "acc_y", 'gy_x', 'gy_z', "gy_y"],
    ["id", "time", "acc_y", "acc_x", "acc_z", 'gy_y', 'gy_x', "gy_z"],
    ["id", "time", "acc_y", "acc_z", "acc_x", 'gy_y', 'gy_z', "gy_x"],
    ["id", "time", "acc_z", "acc_x", "acc_y", 'gy_z', 'gy_x', "gy_y"],
    ["id", "time", "acc_z", "acc_y", "acc_x", 'gy_z', 'gy_y', "gy_x"],
]

realAxis = ["id", "time", "acc_x", "acc_y", "acc_z", 'gy_x', 'gy_y', "gy_z"]

overClassTarget = rawTarget[rawTarget["label_desc"] == "Non-Exercise"]
overClassFeature = makeTimeData(rawFeature.loc[overClassTarget["id"], :], time_length, realAxis[2:])

underClassTarget = rawTarget[rawTarget["label_desc"] != "Non-Exercise"]
underClassFeature = rawFeature.loc[underClassTarget["id"], :]

augData = pd.DataFrame(underClassFeature)
augTarget = pd.DataFrame(underClassTarget)

for axis in tqdm.tqdm(axisComb) :
    rot = underClassFeature[axis]
    rot.columns = realAxis
    rot.loc[:, "id"] += augData["id"].max() + 1
    rot.index = rot["id"]
    augData = pd.concat([augData, rot], axis = 0)
    augTarget = pd.concat([augTarget, underClassTarget], axis = 0)

augData = makeTimeData(augData, time_length, realAxis[2:])

classCount = pd.DataFrame(rawTarget.groupby("label").count()["id"])
classCount.loc[:, "skew"] = (classCount["id"].max() / classCount["id"]).round(decimals = -1).astype(np.int32)
classCount.loc[:, "skew"] = (classCount.loc[:, "skew"] / 6).round(decimals = 0)
augTarget = augTarget.reset_index(drop = True)

tempTargetTable = pd.merge(augTarget.reset_index(), classCount.reset_index().drop("id", axis = "columns")).sort_values("index")
tempTargetTable.index = tempTargetTable["index"]
tempTargetTable = tempTargetTable.drop("index", axis = "columns")

tempAug = augData
tempTarget = augTarget

for ratio in tempTargetTable["skew"].unique() :
    if ratio == 0 :
        next
    for aug in tqdm.tqdm(range(int(ratio))) :
        shift_data = np.roll(tempAug[tempTargetTable[tempTargetTable["skew"] == ratio].index.values], int(random.random()*300), axis = 1)
        augData = np.append(augData, shift_data, axis = 0)
        augTarget = np.append(augTarget, tempTarget.iloc[tempTargetTable[tempTargetTable["skew"] == ratio].index.values, :].values, axis = 0)
        
featureData = np.concatenate([overClassFeature, augData])
targetData = keras.utils.to_categorical(np.concatenate([overClassTarget["label"].values, augTarget[:, 1]]))

trainFeature, testFeature, trainTarget, testTargtet = train_test_split(featureData, targetData, test_size = 0.2)
scaler = StandardScaler()
trainFeature = scaler.fit_transform(trainFeature.reshape(-1, n_feature)).reshape(-1, time_length, n_feature)
testFeature = scaler.transform(testFeature.reshape(-1, n_feature)).reshape(-1, time_length, n_feature)
```

```python
# Augment whole data - class regardless

rawFeature = pd.read_csv("./RawData/train_features.csv")
rawTarget = pd.read_csv("./RawData/train_labels.csv")
rawFeature.index = rawFeature["id"]

overClassTarget = rawTarget[rawTarget["label_desc"] == "Non-Exercise"]
overClassFeature = rawFeature.loc[overClassTarget["id"], :]

underClassTarget = rawTarget[rawTarget["label_desc"] != "Non-Exercise"]
underClassFeature = rawFeature.loc[underClassTarget["id"], :]

axisComb = [
    ["id", "time", "acc_x", "acc_z", "acc_y", 'gy_x', 'gy_z', "gy_y"],
    ["id", "time", "acc_y", "acc_x", "acc_z", 'gy_y', 'gy_x', "gy_z"],
    ["id", "time", "acc_y", "acc_z", "acc_x", 'gy_y', 'gy_z', "gy_x"],
    ["id", "time", "acc_z", "acc_x", "acc_y", 'gy_z', 'gy_x', "gy_y"],
    ["id", "time", "acc_z", "acc_y", "acc_x", 'gy_z', 'gy_y', "gy_x"],
]

realAxis = ["id", "time", "acc_x", "acc_y", "acc_z", 'gy_x', 'gy_y', "gy_z"]

augData = pd.DataFrame(rawFeature)
augTarget = pd.DataFrame(rawTarget)

for axis in tqdm.tqdm(axisComb) :
    rot = rawFeature[axis]
    rot.columns = realAxis
    rot.loc[:, "id"] += augData["id"].max() + 1
    rot.index = rot["id"]
    augData = pd.concat([augData, rot], axis = 0)
    augTarget = pd.concat([augTarget, rawTarget], axis = 0)

time_length = 600
n_feature = 6

def makeTimeData(augmentData, time_length, n_feature) :
    originalFeatureData = np.empty(shape = (augmentData["id"].unique().shape[0], time_length, n_feature))

    i = 0
    for idx in tqdm.tqdm(augmentData.index.unique()) :
        originalFeatureData[i] = augmentData.loc[idx, ["acc_x", "acc_y", "acc_z", "gy_x", "gy_y", "gy_z"]]
        i += 1
    return originalFeatureData
    
featureData = makeTimeData(originalFeatureData, time_length, n_feature)
targetData = pd.DataFrame(augTarget)

for aug in tqdm.tqdm(range(30)) :
    shift_data = np.roll(originalFeatureData, int(random.random()*300), axis = 1)
    featureData = np.append(featureData, shift_data, axis = 0)
    targetData = np.append(targetData, pd.DataFrame(augTarget), axis = 0)
    
trainFeature, testFeature, trainTarget, testTarget = train_test_split(featureData, targetData, test_size = 0.2)
scaler = StandardScaler()
trainFeature = scaler.fit_transform(trainFeature.reshape(-1, n_feature)).reshape(-1, time_length, n_feature)
testFeature = scaler.transform(testFeature.reshape(-1, n_feature)).reshape(-1, time_length, n_feature)
```

```python
class SEBlock(keras.layers.Layer) :
    def __init__(self, reduction_ratio = 16, dense_1_kernel_init = "he_normal", dense_2_kernel_init = "he_normal",
                 dense_1_bias_init = "zeros", dense_2_bias_init = "zeros", **kwargs) :
        
        self.dense_1_kernel = dense_1_kernel_init
        self.dense_2_kernel = dense_2_kernel_init
        self.dense_1_bias = dense_1_bias_init
        self.dense_2_bias = dense_2_bias_init
        self.reduction_ratio = reduction_ratio
        
        super(SEBlock, self).__init__(**kwargs)
        
    def build(self, input_shape) :
        self.w1 = self.add_weight("w1", (input_shape[2], input_shape[2] // self.reduction_ratio),
                                  initializer = keras.initializers.get(self.dense_1_kernel),
                                  trainable = True)
        self.w2 = self.add_weight("w2", (input_shape[2] // self.reduction_ratio, input_shape[2]),
                                  initializer = keras.initializers.get(self.dense_2_kernel),
                                  trainable = True)
        self.b1 = self.add_weight("b1", (input_shape[2] // self.reduction_ratio, ),
                                  initializer = keras.initializers.get(self.dense_1_bias),
                                  trainable = True)
        self.b2 = self.add_weight("b2", (input_shape[2], ),
                                  initializer = keras.initializers.get(self.dense_2_bias),
                                  trainable = True)
        super(SEBlock, self).build(input_shape)
        
    def call(self, x) :
        pool = keras.backend.mean(x, axis = 1, keepdims = True)
        dense_1 = keras.backend.dot(pool, self.w1) + self.b1
        dense_1 = keras.backend.relu(dense_1)
        dense_2 = keras.backend.dot(dense_1, self.w2) + self.b2
        dense_2 = keras.backend.sigmoid(dense_2)
        return dense_2
    
    def compute_output_shape(self, input_shape) :
        return (input_shape[0], input_shape[2])
```

```python
batch_size = 128
n_units = 128

def createModel() :
    InputLayer = keras.layers.Input(shape = (trainFeature.shape[1], trainFeature.shape[2]))

    Shuffle = keras.layers.Permute((2, 1))(InputLayer)
    Attn = keras.layers.AdditiveAttention(use_scale = True)([Shuffle, Shuffle])
    RNN = keras.layers.LSTM(n_units, recurrent_initializer = "glorot_normal", dropout = 0.5)(Attn)
    
    Conv = keras.layers.Conv1D(filters = n_units, kernel_size = 8, strides = 1, padding = "same", activation = None, kernel_initializer = "he_uniform")(InputLayer)
    Norm = keras.layers.BatchNormalization(momentum = 0.99, epsilon = 0.001)(Conv)
    Activ = keras.layers.Activation("relu")(Norm)
    Drop = keras.layers.Dropout(rate = 0.5)(Activ)
    Block = SEBlock(reduction_ratio = 8)(Drop)
    Res = keras.layers.Multiply()([Drop, Block])
    
    Conv = keras.layers.Conv1D(filters = n_units * 2, kernel_size = 5, strides = 1, padding = "same", activation = None, kernel_initializer = "he_uniform")(Res)
    Norm = keras.layers.BatchNormalization(momentum = 0.99, epsilon = 0.001)(Conv)
    Activ = keras.layers.Activation("relu")(Norm)
    Drop = keras.layers.Dropout(rate = 0.5)(Activ)
    Block = SEBlock(reduction_ratio = 8)(Drop)
    Res = keras.layers.Multiply()([Drop, Block])
    
    Conv = keras.layers.Conv1D(filters = n_units, kernel_size = 3, strides = 1, padding = "same", activation = None, kernel_initializer = "he_uniform")(Res)
    Norm = keras.layers.BatchNormalization(momentum = 0.99, epsilon = 0.001)(Conv)
    Activ = keras.layers.Activation("relu")(Norm)
    Drop = keras.layers.Dropout(rate = 0.5)(Activ)
    Pool = keras.layers.GlobalAveragePooling1D()(Drop)
    
    Concat = keras.layers.Concatenate()([RNN, Pool])
    
    OutputLayer = keras.layers.Dense(trainTarget.shape[1], activation = "softmax")(Concat)
    
    model = keras.models.Model(inputs = InputLayer, outputs = OutputLayer)
    optimizer = keras.optimizers.SGD()
    model.compile(optimizer = optimizer, loss = keras.losses.CategoricalCrossentropy(), metrics = ["accuracy"])
    return model

model = createModel()
model.summary()
```

```python
es = keras.callbacks.EarlyStopping(monitor = "val_loss", min_delta = 0.0001, patience = 30, restore_best_weights = True)
lr = keras.callbacks.ReduceLROnPlateau(monitor = "val_loss", factor = np.log10(3), patience = 5, verbose = 1, min_delta = 0.0001)

model = createModel()
history = model.fit(trainFeature, trainTarget,
                    epochs = 1000, validation_split = 0.35, batch_size = batch_size, callbacks = [lr, es])
```
